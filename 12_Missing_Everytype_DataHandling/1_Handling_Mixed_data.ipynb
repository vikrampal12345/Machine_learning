```python
import pandas as pd
import numpy as np
import seaborn as sns

df = pd.read_csv('/content/titanic.csv', usecols=['Ticket', 'Cabin', 'number','Survived'])

df.sample(5)
```

```python
df['number'].unique()
```

```python
df['number'].value_counts().plot(kind='bar')
```

```python
# Extract the numerical Data
# pd.to_numeric(data, errors=..., downcast=...)
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
```

```python
# extract the categorical Data

df['number_categorical'] = np.where(df['number_numerical'].isnull(), df['number'], np.nan)

df.head()
```

```python
df['Cabin'].unique()
```

```python
df['cabin_num'] = df['Cabin'].str.extract('(\d+)')  # captures numerical part
df['cabin_num']
```

```python
df['cabin_cat'] = df['Cabin'].str[0]  #capture the first letter
df['cabin_cat']
```

```python
df.sample(5)
```

```python
df['Ticket'].unique()
```

```python
# extract the last bit of ticket as number

df['ticket_num'] = df['Ticket'].apply(lambda s:s.split()[-1]) # captures numerical part
df['ticket_num'] = pd.to_numeric(df['ticket_num'], errors = 'coerce', downcast='integer')


# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda s:s.split()[0])  #capture the first letter
df['ticket_cat'] = np.where(df['ticket_cat'].str.isdigit(), np.nan, df['ticket_cat'])
```
