# Handle Date and Time columns

```python
import pandas as pd
import numpy as np

date = pd.read_csv("orders.csv")
time = pd.read_csv("messages.csv")
```

## working with date


```python
# Converting to datetime datatype

date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract year


```python
date['date_year'] = date['date'].dt.year
date.head()
```

2. Extract the month

```python
date['date_month_no'] = date['date'].dt.month
date.head()
```

3. Extract the day


```python
date['date_day'] = date['date'].dt.day
date.head()
```

4. Extract the month name


```python
date['date_month_name'] = date['date'].dt.month_name()
date.head()
```

5. day of week

```python
date['date_dow'] = date['date'].dt.dayofweek
date.head(2)
```

6. Extract the day name

```python
date['date_day_week_name'] = date['date'].dt.day_name()
date.head()
```

7. Extract the weekend

```python
date['date_weekend'] = np.where(date['date_day_week_name'].isin(['Sunday', 'Saturday']),1,0)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

8. Extract week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

9. Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

10. Extract the semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]),1,2)
date.drop(columns=['product_id', 'city_id', 'orders']).head()
```

11. Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
# day
(today - date['date']).dt.days
```

```python
# months passed

import pandas as pd

today = pd.Timestamp.today()

date['date'] = pd.to_datetime(date['date'])

date['months_passed'] = (
    (today.year - date['date'].dt.year) * 12
    + (today.month - date['date'].dt.month)
)
date.head()
```
