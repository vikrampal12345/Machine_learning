# Simple Linear Regression (Mathmetical Formulation)

```python
# two way to find the solution
# 1. Closed form solution (OLS) - LinearRegression
# 2. Non-closed form solution (Gradient descent) - SUDRegression
```

```python
class MeraLR:

  def __init__(self):
    self.m = None
    self.b = None

  def fit(self, X_train, y_train):
    num = 0
    deno = 0

    for i in range(X_train.shape[0]):
      num = num +  (X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean())
      deno = deno + (X_train[i] - X_train.mean()) * (X_train[i] - X_train.mean())

    self.m = num/deno
    self.b = y_train.mean() - (self.m * X_train.mean())

    print(self.m)
    print(self.b)
  def predict(self, X_test):
    return self.m * X_test + self.b
```
