# Feature Construction

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
```

```python
dff = sns.load_dataset('titanic')
df = dff.iloc[:,[0,1,3,4,5]]
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
x = df.drop('survived', axis=1)
y = df.iloc[:,0]
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
np.mean(cross_val_score(LogisticRegression(), x, y, scoring='accuracy', cv=20))
```

# Applying Feature Construction

```python
x["Family_size"] = x['sibsp'] + x['parch'] + 1  # passagner also consider
```

```python
x
```

```python
def myfunc(num):
  if num == 1:
    return 0
  elif num > 1 and num <= 4:
    return 1
  else:
    return 2

```

```python
x['Family_type'] = x['Family_size'].apply(myfunc)
```

```python
x
```

```python
x = x.iloc[:,[0,1,5]]
x
```

```python
np.mean(cross_val_score(LogisticRegression(), x, y, scoring='accuracy', cv=20))
```

# Feature Splitting

```python
var = pd.read_csv('/content/Tatanic_dataset.csv')
var.head()
```

```python
var['Title'] = var['Name'].str.split(', ', expand=True)[1].str.split('.', expand=True)[0]
```

```python
var
```

```python
var.groupby('Title')['Survived'].mean().sort_values(ascending=False)
```

```python
var['Is_married'] = 0
var['Is_married'].loc[var['Title'] == 'Mrs'] = 1
```
