# Exploratory Data Analysis (EDA)

## Full Form

EDA = **Exploratory Data Analysis**

EDA is the process of understanding, cleaning, summarizing, and visualizing data before applying machine learning models.

---

# Types of EDA

EDA is mainly divided into **4 types**:

## 1. Univariate Analysis

### Definition

Analysis of a single variable at a time.

### Purpose

* Understand distribution
* Find outliers
* Find central tendency

### Examples

* Mean
* Median
* Mode
* Histogram
* Box Plot

```python
df['Age'].describe()
```

---

## 2. Bivariate Analysis

### Definition

Analysis of two variables together.

### Purpose

* Find relationships
* Compare variables
* Detect trends

### Examples

* Correlation
* Scatter Plot
* Group Comparisons

```python
df['Age'].corr(df['Salary'])
```

---

## 3. Multivariate Analysis

### Definition

Analysis of more than two variables simultaneously.

### Purpose

* Discover complex relationships
* Feature selection
* Pattern detection

### Examples

* Correlation Matrix
* Pair Plot
* Heatmap

```python
df.corr(numeric_only=True)
```

---

## 4. Data Cleaning Analysis

### Definition

Checking and fixing data quality issues.

### Purpose

* Remove errors
* Improve dataset quality

### Tasks

* Missing Values
* Duplicate Values
* Incorrect Data Types
* Outliers

```python
df.isnull().sum()
df.duplicated().sum()
```

# Types of Data

Data is mainly divided into **2 categories**:

## 1. Qualitative Data (Categorical Data)

This data describes qualities or characteristics and cannot be measured numerically.

### Types

### A. Nominal Data

Categories with no specific order.

Examples:

* Gender (Male, Female)
* City (Delhi, Mumbai, Pune)
* Color (Red, Blue, Green)

### B. Ordinal Data

Categories with a meaningful order.

Examples:

* Education Level (High School, Bachelor's, Master's)
* Customer Rating (Poor, Good, Excellent)
* T-Shirt Size (S, M, L, XL)

---

## 2. Quantitative Data (Numerical Data)

This data consists of numbers and can be measured.

### Types

### A. Discrete Data

Countable values.

Examples:

* Number of Students
* Number of Employees
* Number of Products Sold

```python
1, 2, 3, 4, 5
```

### B. Continuous Data

Measurable values that can contain decimals.

Examples:

* Height
* Weight
* Temperature
* Salary

```python
65.5, 72.3, 180.75
```

# Box Plot (Box-and-Whisker Plot)

## What is a Box Plot?

A Box Plot is a graphical representation used to visualize the distribution of numerical data and detect outliers.

It summarizes the dataset using:

* Minimum Value
* First Quartile (Q1)
* Median (Q2)
* Third Quartile (Q3)
* Maximum Value

---

## Structure of a Box Plot

```text
Outlier    Min      Q1      Median      Q3      Max
   *        |--------|====|====|--------|
            Whisker      Box       Whisker
```

### Components

### 1. Minimum Value

The smallest value excluding outliers.

### 2. First Quartile (Q1)

25% of data lies below this value.

### 3. Median (Q2)

50% of data lies below this value.

### 4. Third Quartile (Q3)

75% of data lies below this value.

### 5. Maximum Value

Largest value excluding outliers.

### 6. Outliers

Values that are unusually high or low compared to the rest of the data.

---

## How Box Plot Detects Outliers

### Step 1: Calculate IQR

IQR = Q_3 - Q_1

### Step 2: Calculate Lower and Upper Limits

Lower Limit:

```text
Q1 - 1.5 × IQR
```

Upper Limit:

```text
Q3 + 1.5 × IQR
```

### Step 3: Identify Outliers

Any value:

* Below Lower Limit
* Above Upper Limit

is considered an outlier.

---

## Example

Dataset:

```python
[10, 12, 14, 15, 18, 20, 22, 100]
```

Here:

* Most values are between 10 and 22.
* 100 is far away from the rest.

The box plot will show:

* Main data inside the box.
* 100 as an outlier point.

---

## Why Use Box Plots?

### Advantages

* Detect outliers quickly.
* Understand data spread.
* Compare multiple distributions.
* Identify skewness.
* Visualize quartiles and median.

---

```python

```

---

## Interpretation

### Symmetric Data

* Median is near the center of the box.

### Right-Skewed Data

* Upper whisker is longer.
* Median closer to Q1.

### Left-Skewed Data

* Lower whisker is longer.
* Median closer to Q3.

### Presence of Outliers

* Dots outside whiskers indicate outliers.
