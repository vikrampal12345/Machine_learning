```python
# ordinal data - ordinal encoding - input data
# nominal data - one hot encoding
# output data - label encoding

import pandas as pd
import numpy as np

df = pd.read_csv("/content/customer.csv")
df.head()
```

```python
df = df.iloc[:,2:]
df
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.drop('purchased', axis=1), df['purchased'], test_size=0.2 )

X_train
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['Poor', 'Average', 'Good'], ['School', 'UG', 'PG']])

oe.fit(X_train)
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```

```python
X_train
```

```python
oe.categories_        # check categories and their datatype
```
