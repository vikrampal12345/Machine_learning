# Correlation

## What is Correlation?

Correlation measures the strength and direction of the relationship between two numerical variables.

It helps us understand how one variable changes when another variable changes.

For example:

* If company ratings increase and employee satisfaction also increases, they have a positive correlation.
* If product price increases and sales decrease, they have a negative correlation.

## Correlation Coefficient

The correlation coefficient ranges from -1 to +1.

| Correlation Value | Meaning                       |
| ----------------- | ----------------------------- |
| +1                | Perfect Positive Correlation  |
| 0.7 to 0.99       | Strong Positive Correlation   |
| 0.3 to 0.69       | Moderate Positive Correlation |
| 0                 | No Correlation                |
| -0.3 to -0.69     | Moderate Negative Correlation |
| -0.7 to -0.99     | Strong Negative Correlation   |
| -1                | Perfect Negative Correlation  |

## Types of Correlation

### 1. Positive Correlation

When one variable increases, the other variable also increases.

Example:

* Experience ↑ → Salary ↑

### 2. Negative Correlation

When one variable increases, the other variable decreases.

Example:

* Product Price ↑ → Sales ↓

### 3. No Correlation

No relationship exists between the variables.

Example:

* Employee ID and Salary

## Correlation Formula

r=\frac{\sum (x_i-\bar{x})(y_i-\bar{y})}{\sqrt{\sum (x_i-\bar{x})^2\sum (y_i-\bar{y})^2}}

Where:

* r = Correlation coefficient
* x = First variable
* y = Second variable
* x̄ = Mean of x
* ȳ = Mean of y

## Correlation in Pandas

Calculate correlation matrix:

```python

```

Calculate correlation between two columns:

```python

```

## Heatmap Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.corr(numeric_only=True),
            annot=True)

plt.show()
```

## Important Notes

* Correlation only works on numerical data.
* Correlation does not imply causation.
* A high correlation does not mean one variable causes the other.
* Values close to +1 or -1 indicate a strong relationship.
* Values close to 0 indicate a weak or no relationship.
