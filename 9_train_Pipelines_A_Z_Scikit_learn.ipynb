# creating pipeline

````python
# imputation transformer (Nan replace by the value)

trf1 = ColumnTransformer([
    ('impute_age', SimpleImputer(),[2]),
     ('impute_embarked', SimpleImputer(strategy='most_frequent'),[6])# use index in the pipeline,
], remainder='passthrough')             # because it return the numpy array and the no dataframe. no dataframe means no column
````

````python
# one hot encoding
trf2 = ColumnTransformer([
    ('ohe_sex_embarked', OneHotEncoder(sparse_output=False, handle_unknown='ignore'),[1,6])
], remainder='passthrough')
````

````python
# scaling
trf3 = ColumnTransformer(transformers=[
    ('scale', MinMaxScaler(), slice(0, 10))
    ])
````

````python
# Feature selection
from sklearn.feature_selection import SelectKBest, chi2
trf4 = SelectKBest(score_func=chi2, k=8)
````

````python
# train the model
trf5 = DecisionTreeClassifier()
````

# creating pipeline here

````python
pipe = Pipeline([
    ('trf1', trf1),
    ('trf2', trf2),
    ('trf3', trf3),
    ('trf4', trf4),
    ('trf5', trf5)
    ])
````

## OneHotEncoder Parameters – Short Notes

### `sparse_output=False`

* Converts the encoded output into a **NumPy array**.
* Default output is a **sparse matrix** to save memory.
* Useful for **small datasets** and easier visualization.

```python
OneHotEncoder(sparse_output=False)
```

**Output Type:**

* `False` → NumPy array (`ndarray`)
* `True` (default) → Sparse matrix (`csr_matrix`)

---

### `drop='first'`

* Removes the first category after one-hot encoding.
* Prevents the **dummy variable trap** (multicollinearity).
* Reduces redundant columns.

```python
OneHotEncoder(drop='first')
```

**Example:**

Categories: `['Red', 'Blue', 'Green']`

Without `drop='first'`:

| Blue | Green | Red |
| ---- | ----- | --- |
| 1    | 0     | 0   |
| 0    | 1     | 0   |
| 0    | 0     | 1   |

With `drop='first'`:

| Green | Red |         |
| ----- | --- | ------- |
| 0     | 0   | ← Blue  |
| 1     | 0   | ← Green |
| 0     | 1   | ← Red   |

---

### Important Note

* These parameters are used during **encoding**, not **scaling**.
* Scaling is performed using tools like:

  * `StandardScaler`
  * `MinMaxScaler`
  * `RobustScaler`


## Pipeline + ColumnTransformer – Notes

### Why Use a Pipeline?

A `Pipeline` applies multiple preprocessing steps **sequentially** to the same column.

Example:

```text id="j3c1y4"
embarked
   ↓
SimpleImputer
   ↓
OneHotEncoder
```

Without a `Pipeline`, `ColumnTransformer` applies transformations independently.

---

### Required Imports

```python id="krm9q7"
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
```

---

### Create a Pipeline for `embarked`

```python id="n6f8w2"
embarked_pipe = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(drop='first', sparse_output=False))
])
```

#### Why these steps?

* `SimpleImputer(strategy='most_frequent')`

  * Fills missing values with the most common category.
  * Example: `S, C, NaN, S → S, C, S, S`

* `OneHotEncoder(drop='first', sparse_output=False)`

  * Converts categories into numerical columns.
  * `drop='first'` avoids the dummy variable trap.
  * `sparse_output=False` returns a NumPy array instead of a sparse matrix.

---

### Create the ColumnTransformer

```python id="z5d1a8"
transformer = ColumnTransformer([
    ('age_imputer', SimpleImputer(), ['age']),

    ('sex_encoder',
     OneHotEncoder(drop='first', sparse_output=False),
     ['sex']),

    ('embarked_pipeline',
     embarked_pipe,
     ['embarked'])
], remainder='passthrough')
```

---

### Fit and Transform

```python id="m7r4x9"
X_train_transformed = transformer.fit_transform(X_train)
X_test_transformed = transformer.transform(X_test)
```

---

## Why Use These Important Steps?

| Step                  | Purpose                                              |
| --------------------- | ---------------------------------------------------- |
| `SimpleImputer()`     | Fills missing values                                 |
| `OneHotEncoder()`     | Converts categorical data to numbers                 |
| `drop='first'`        | Prevents multicollinearity                           |
| `sparse_output=False` | Returns a normal NumPy array                         |
| `Pipeline()`          | Applies multiple steps to the same column            |
| `ColumnTransformer()` | Applies different preprocessing to different columns |
| `fit_transform()`     | Learns parameters and transforms training data       |
| `transform()`         | Applies learned parameters to test data              |

---

## Remember

```text id="b9k2u6"
Pipeline → Same column, multiple steps

ColumnTransformer → Different columns, different transformations
```
