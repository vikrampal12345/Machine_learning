# n-1 Encoding ( remove the first column so the missilinity not arise(means every input column are mathmatically independent))

```python
# we cannot use the pandas onehot encoding in the machine learning
# because it not remember the columns.
# so everytime it give different value

pd.get_dummies(df, columns = ['fuel', 'owner'], drop_first=True)
```

# 4. OneHotEncoding with Top Categories

```python
counts = df["brand"].value_counts() # count the value of the brand
```

```python
df['brand'].nunique()
threshold = 100
```

```python
relp = counts[counts <= threshold].index  # we extarct the brand which is less than 100 cars
```

```python
pd.get_dummies(df['brand'].replace(relp, 'uncommon')).sample(5)
```
